# 分析目标

此数据分析目的是，根据市场销售数据，**挖掘畅销产品**，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。




# 简介

原始数据记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：

 - InvoiceNo:发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消
 - StockCode:产品代码。5位数，作为产品的唯一标识符。
 - Description:产品名称。
 - Quantity:产品在交易中的数量。
 - InvoiceDate:发票日期和时间。交易发生的日期和时间。
 - UnitPrice:单价。价格单位为英镑。
 - CustomerID:客户编号。5位数，作为客户的唯一标识符。
 - Country:国家名称，客户所居住的国家的名称。

### 评估无效或错误数据

可以通过describe方法，对数值统计信息快速了解

```python
df.describe()
```

```python
#通过观察发现CustomerID(客户编号)列被数值计算了，说明是数值型数据； 客户编号不应该是数值型而是字符串型数据
#Quantity(商品交易数量)和UnitPrice(商品价格)都有负数； 按道理商品交易数量和商品价格不应该有负数
```

```python
#先把Quantity为负数的数据筛选出来，评估下意义
```

```python
df[df['Quantity']<0]
```

#### 观察发现可以猜测Quantity为负数时，InvoiceNo(订单ID)是C开头，而订单ID是C开头代表“交易取消”，符合商品交易数量为负数，
#### 但中间省略了很多行，所以在Quantity为负数的基础上多添加一个条件InvoiceNO开头不以C开头进行筛选，如果没有则猜测正确

```python
df[(df['Quantity'] < 0) & (df['InvoiceNo'].str[0] != 'C')]
```

#### 观察发现当Quantity为负数时，InvoiceNo有些不是'C’开头，所以猜测错误，存在有一些不是退货的这个交易，
#### 但UnitPrice都是0
#### 所以再加一个条件进行筛选，在Quantity为负数、InvoiceNo不以'C’开头的基础上是否存在UnitPrice不为0的数据

```python
df[(df['Quantity'] < 0) & (df['InvoiceNo'].str[0] != 'C') & (df['UnitPrice'] != 0)]
```

#### 观察发现在在Quantity为负数、InvoiceNo不以'C’开头的基础上UnitPrice不存在不为0的情况

#### 结论：当Quantity为负数时都满足以下条件之一：
- InvoiceNo 以C开头，订单被取消
- UnitPrice 为0，说明单价为0英镑
#### 这些数据都不是有效成交数据，不贡献销售，对分析畅销商品没帮助
#### 而且会影响数值分析，所以将Quantity为负数的数据全部删除

#### 接下来筛选出UnitPrice为负数的观测值

```python
df[df['UnitPrice']<0]
```

#### 观察发现UnitPrice(商品单价)为0时，Description(商品名称）都为Adjust bad debt（坏账调整），说明不是实际的商品交易数据，可以删除

# 清洗数据

根据前面的评估得到的结论，我们需要进行的数据清洗包括：
- 把InvoiceDate 变量的数据类型转换为日期时间
- 把CustomerID 变量的数据类型转换为字符串
- 把Description 变量缺失的观测值删除
- 把Quantity 变量值为负数的观测值删除
- 把UnitPrice 变量值为负数的观测值删除


为了区分开经过清洗的数据和原始的数据，我们创建新的变量cleaned_data，让它为df复制出的副本，我们之后的清理步骤都将被运用在cleaned_data上。

```python
cleaned_data=df.copy()
```

```python
cleaned_data.head()
```

#### 把InvoiceDate 变量的数据类型转换为日期时间

```python
pd.to_datetime(cleaned_data['InvoiceDate'])
```

```python
cleaned_data['InvoiceDate']=pd.to_datetime(cleaned_data['InvoiceDate'])
cleaned_data['InvoiceDate']
```

#### 把CustomerID 变量的数据类型转换为字符串

```python
cleaned_data['CustomerID']=cleaned_data['CustomerID'].astype(str)
cleaned_data['CustomerID']
```

#### 由于之前CustomerID（客户编号）为数值型，转换成字符串带小数点，但CustomerID（客户编号）不要带小数点
#### 把CustomerID变量值结尾的 ‘.0’删除

```python
# slice()方法，保留字符串的一部分  参数1 起始索引 , 参数2 结束位置的下一个索引  (包头不包尾)
```

```python
cleaned_data['CustomerID'].str.slice(0,5)
```

```python
cleaned_data['CustomerID']=cleaned_data['CustomerID'].str.slice(0,5)
cleaned_data['CustomerID']
```

#### 把Description 变量缺失的观测值删除

```python
cleaned_data=cleaned_data.dropna(subset=['Description'])
cleaned_data
```

```python
cleaned_data['Description'].isnull().sum()
```

#### 把Quantity 变量值为负数的观测值删除

```python
cleaned_data=cleaned_data[cleaned_data['Quantity'] >= 0]
cleaned_data
```

```python
cleaned_data[cleaned_data['Quantity'] < 0]
```

#### 把UnitPrice 变量值为负数的观测值删除

```python
cleaned_data=cleaned_data[cleaned_data['UnitPrice'] >= 0]
cleaned_data
```

```python
cleaned_data[cleaned_data['UnitPrice'] < 0]
```

# 保存好清理后的数据

```python
cleaned_data.to_csv('cleaned_data.csv',index=False) #数据的索引为位置索引，没有意义，不想保存成新的一列，index=False
```
